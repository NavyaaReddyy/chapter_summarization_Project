```python
import nltk
import spacy
from transformers import BartForConditionalGeneration, BartTokenizer
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize

def preprocess_text(text):
    sentences = sent_tokenize(text)
    return sentences


chapter_text = """
In the beginning, there was darkness. Then, the universe was created with a big bang.
Over billions of years, galaxies formed, stars ignited, and planets coalesced.
On one such planet, life emerged, evolving from simple organisms to complex beings capable of consciousness.
Humanity rose, built civilizations, and began to explore the mysteries of the cosmos.
They wondered about their place in the universe, about the nature of existence itself.
"""

preprocessed_sentences = preprocess_text(chapter_text)
print(preprocessed_sentences)
```

```python
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')
```

```python
def summarize_text(text, max_length=130, min_length=30):
    inputs = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=1024, truncation=True)
    summary_ids = model.generate(inputs, max_length=max_length, min_length=min_length, length_penalty=2.0, num_beams=4, early_stopping=True)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

summary = summarize_text(chapter_text)
print("Summary:")
print(summary)
```
